#### Задача 4

```python
M4_x = 174
sigma = 8
def countsigma(x):
    return float((x - M4_x) / sigma)
```

```python
# а). Вероятность того, что рост больше 182 см:
Z_a = countsigma(182)    # При 1sigma вероятность - 0.8413 (таблица)
P_a = 1 - 0.8413
print(f'Количество сигм равно: {Z_a:.2f}')
print(f'Вероятность равна: {P_a:.4f} или {P_a*100:.2f} %')
```

```python
# б). Вероятность того, что рост больше 190 см:
Z_б = countsigma(190)    # При 2sigma вероятность - 0.9772 (таблица)
P_б = 1 - 0.9772
print(f'Количество сигм равно: {Z_б:.2f}')
print(f'Вероятность равна: {P_б:.4f} или {P_б*100:.2f} %')
```

```python
# в). Вероятность того, что рост от 166 см до 190 см:
Z_в1 = countsigma(166)    # При -1sigma вероятность - 0.1587 (таблица)
Z_в2 = countsigma(190)    # При 2sigma вероятность - 0.9772 (таблица)
P_в = 0.9772 - 0.1587    
print(f'Количество сигм при росте 166 см равно {Z_в1:.2f}, при росте 190 см равно {Z_в2:.2f}')
print(f'Вероятность равна: {P_в:.4f} или {P_в*100:.2f} %')
```

```python
# г). Вероятность того, что рост от 166 см до 182 см:
Z_г1 = countsigma(166)    # При -1sigma вероятность - 0.1587 (таблица)
Z_г2 = countsigma(182)    # При 1sigma вероятность - 0.8413 (таблица)
P_г = 0.8413 - 0.1587    
print(f'Количество сигм при росте 166 см равно {Z_г1:.2f}, при росте 182 см равно {Z_г2:.2f}')
print(f'Вероятность равна: {P_г:.4f} или {P_г*100:.2f} %')
```

```python
# д). Вероятность того, что рост от 158 см до 190 см:
Z_д1 = countsigma(158)    # При -2sigma вероятность - 0.0228 (таблица)
Z_д2 = countsigma(190)    # При 2sigma вероятность - 0.9772 (таблица)
P_д = 0.9772 - 0.0228    
print(f'Количество сигм при росте 158 см равно {Z_д1:.2f}, при росте 190 см равно {Z_д2:.2f}')
print(f'Вероятность равна: {P_д:.4f} или {P_д*100:.2f} %')
```

```python
# е). Вероятность того, что рост не выше 150 см или не ниже 190 см:
Z_е1 = countsigma(150)    # При -3sigma вероятность - 0.0014 (таблица)
Z_е2 = countsigma(190)    # При 2sigma вероятность - 1-0.9772 (таблица)
P_е = 0.0014 + 0.0228    
print(f'Количество сигм при росте 150 см равно {Z_е1:.2f}, при росте 190 см равно {Z_е2:.2f}')
print(f'Вероятность равна: {P_е:.4f} или {P_е*100:.2f} %')
```

```python
# ё). Вероятность того, что рост не выше 150 см или не ниже 198 см:
Z_ё1 = countsigma(150)    # При -3sigma вероятность - 0.0014 (таблица)
Z_ё2 = countsigma(198)    # При 3sigma вероятность - 1-0.9986 (таблица)
P_ё = 0.0014 + 0.0014    
print(f'Количество сигм при росте 150 см равно {Z_ё1:.2f}, при росте 198 см равно {Z_ё2:.2f}')
print(f'Вероятность равна: {P_ё:.4f} или {P_ё*100:.2f} %')
```

```python
# ж). Вероятность того, что рост ниже 166 см:
Z_ж = countsigma(166)    # При -1sigma вероятность - 0.1587 (таблица)
P_ж = 0.1587
print(f'Количество сигм равно: {Z_ж:.2f}')
print(f'Вероятность равна: {P_ж:.4f} или {P_ж*100:.2f} %')
```
